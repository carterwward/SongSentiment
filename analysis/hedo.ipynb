```python
hedonometer = pd.read_csv("data_and_models/Hedonometer.csv")
```

```python
word_valence_dict = {row['Word']: row['Happiness Score'] for index, row in hedonometer.iterrows()}
```

```python
model = pickle.load(open('data_and_models/model.sav', 'rb'))
vectorizer = pickle.load(open('data_and_models/vectorizer.pk', 'rb')) 
```

```python
model_coef = ((model.coef_).transpose())
vocab = vectorizer.get_feature_names()
df = pd.DataFrame(model_coef, columns=["Model Coefficient"])
feature = pd.DataFrame(vocab, columns = ["Feature"])
feature = pd.concat([feature, df], axis=1)
feature = feature.set_index('Feature')
```

```python
missing_features = list(set(vocab) - set(hedonometer['Word'].values))
feature = feature.drop(missing_features)
```

```python
hedo_scores = [word_valence_dict[word] for word in feature.index]
feature['Happiness Score'] = hedo_scores
```

```python
feature = feature.reset_index()
feature
```

```python
fig = px.scatter(data_frame=feature, x='Model Coefficient', y='Happiness Score', hover_data=['Feature'])
```
