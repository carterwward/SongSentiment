```python
import pandas as pd
import pickle as pk
import numpy as np
import os
from twitter_bot.firebase import read_artist_dict
```

```python
train = pd.read_csv("../data_and_models/train_data.csv")
test = pd.read_csv("../data_and_models/test_data.csv")
```

```python
train_lyrics = train['tokenized_lyrics'].values
test_lyrics = test['tokenized_lyrics'].values
tokenized_lyrics = np.concatenate((train_lyrics,test_lyrics))
```

```python
print(tokenized_lyrics)
```

```python
tokenized_lists = [token_string.split(' ') for token_string in tokenized_lyrics]
```

```python
tokenized_lists
```

```python
tokenized_words = set([item for sublist in tokenized_lists for item in sublist])
```

```python
len(tokenized_words)
```

```python
intersection = hedonometer_set&tokenized_words
print(len(intersection))
```

```python
tokenized_words-intersection
```
