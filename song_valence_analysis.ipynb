```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train_data.csv")
test = pd.read_csv("test_data.csv")
```

```python
valences = list(test['valence'].values) + list(train['valence'].values)
```

```python
valence_hist = sns.displot(valences, kde=True)
plt.title('Distribution of Song Valence Scores')
plt.xlabel('Valence Score')
valence_hist.savefig("valence_hist.png")
```

```python
lr_values = ['positive' if val > 0.5 else 'negative' for val in valences]
```

```python
cat_barplot = sns.displot(lr_values, shrink=.8)
plt.title('Number of Positive and Negative Songs')
plt.xlabel('General Sentiment')
cat_barplot.savefig("positive_vs_negative_plot.png")
```
